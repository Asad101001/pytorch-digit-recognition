```python
# Cell 1: Import Libraries
import torch
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Explanation: PIL for images, matplotlib for visualization.
```

```python
# Cell 2: NumPy to Tensor and Back
np_array = np.array([1, 2, 3, 4])
torch_tensor = torch.from_numpy(np_array)  # Convert NumPy to Tensor
print("NumPy Array:", np_array)
print("Torch Tensor:", torch_tensor)

# Back to NumPy
back_to_np = torch_tensor.numpy()
print("Back to NumPy:", back_to_np)

# Explanation: Shared memoryâ€”changes in one affect the other. Useful for data loading.
```

```python
# Cell 3: List to Tensor
list_data = [1, 2, 3]
tensor_from_list = torch.tensor(list_data)
print("Tensor from List:", tensor_from_list)

# Explanation: Direct conversion for simple data.
```

```python
# Cell 4: Image to Tensor (Simulated)
# Create a fake image (or load one with Image.open('path.jpg'))
fake_image = np.random.rand(28, 28, 3) * 255  # 28x28 RGB image
pil_image = Image.fromarray(fake_image.astype('uint8'))
tensor_image = torch.from_numpy(np.array(pil_image)).permute(2, 0, 1).float() / 255  # Channels first, normalize
print("Image Tensor Shape:", tensor_image.shape)  # Should be [3, 28, 28]

# Visualize
plt.imshow(pil_image)
plt.title("Fake Image")
plt.show()

# Explanation: Images are arrays. Permute rearranges to PyTorch format (channels, height, width). Normalize to [0,1].
```

```python
# Cell 5: Data Types and Devices
tensor_int = torch.tensor([1, 2, 3], dtype=torch.int32)
tensor_float = tensor_int.to(torch.float32)  # Change type
print("Float Tensor:", tensor_float)

# Move to device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
tensor_device = tensor_float.to(device)
print("Tensor on Device:", tensor_device.device)

# Explanation: Types matter (e.g., float for images). Devices: CPU or GPU.
```
